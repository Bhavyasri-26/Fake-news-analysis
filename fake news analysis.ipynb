```python
#replace null values with empty string
data=data.fillna('')
```

```python
X=data.drop(columns='label', axis=1)
Y=data['label']
```

Stemming - reducing to root word

```python
port_stem=PorterStemmer()
```

```python
def stemming(content):  #stemming-inbuit python function
  s_content=re.sub('[^a-zA-Z]',' ',content)
  s_content=s_content.lower()
  s_content=s_content.split()
  s_content=[port_stem.stem(word) for word in s_content if not word in stopwords.words('english')]
  s_content=' '.join(s_content)
  return s_content
```

```python
data['content']=data['content'].apply(stemming)
```

```python
data['content']
```

```python
#seperating data and label
X=data['content'].values
Y=data['label'].values
```

```python
X
```

```python
Y
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
#converting textual data to numerical data
vectorizer=TfidfVectorizer()
vectorizer.fit(X)
X=vectorizer.transform(X)
```

```python
print(X)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#splitting dataset to training & testing
X_train, X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y, random_state=2)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
#accuracy
X_train_prediction=model.predict(X_train)
accuracy=accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy score in percentage: ",accuracy*100)
```

```python
X_new=X_test[3256]
prediction=model.predict(X_new)
print(prediction)

if(prediction[0]==0):
  print("The news is REAL")
else:
  print("Te news is FAKE")
```
